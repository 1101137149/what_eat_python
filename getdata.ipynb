```python
import requests
import csv
import io

#轉換CSV資料取得某欄資料
def csvdata(URL,rowNum,dataType):
    reponse=requests.get(URL) #抓網址
    reponse.encoding='utf-8' #轉編碼
    csv_txt=reponse.text
    file=io.StringIO(csv_txt) #將變數輸出成string型態
    csv_reader=csv.reader(file) #reader 只能讀一次 之後會消失
    csv_list=list(csv_reader) #把資料轉成list型態

    if dataType ==1:
#print(csv_list)
        data=set()
        for row in csv_list:
            data.add(row[rowNum])
        print(data)
    else:
        data=set()
        for row in csv_list:
            data.add(f"{row[rowNum-1]}_{row[rowNum]}")
        print(data)

#城市
city=csvdata('https://quality.data.gov.tw/dq_download_csv.php?nid=101905&md5_url=ac4416b24d75780b583912a43905a526',1,1)
#城市_區域
area=csvdata('https://quality.data.gov.tw/dq_download_csv.php?nid=5948&md5_url=b7ce2082883f6b680810828799d4c32e',2,2)
```
